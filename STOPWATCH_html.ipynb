```python
<!DOCTYPE html>
<html lang="en">
<head>
  <meta charset="UTF-8">
  <title>Stopwatch App</title>
  <style>
    body {
      font-family: 'Segoe UI', sans-serif;
      display: flex;
      flex-direction: column;
      align-items: center;
      justify-content: center;
      height: 100vh;
      margin: 0;
      background: #f4f4f4;
    }

    .stopwatch {
      background: #fff;
      padding: 30px;
      border-radius: 15px;
      box-shadow: 0 0 15px rgba(0,0,0,0.2);
      text-align: center;
      width: 300px;
    }

    h1 {
      margin-top: 0;
      font-size: 2em;
      color: #333;
    }

    #display {
      font-size: 2.5em;
      margin: 20px 0;
    }

    .buttons button {
      padding: 10px 20px;
      margin: 5px;
      font-size: 1em;
      border: none;
      border-radius: 5px;
      cursor: pointer;
      transition: 0.2s;
    }

    .start { background: #28a745; color: white; }
    .pause { background: #ffc107; color: white; }
    .reset { background: #dc3545; color: white; }
    .lap   { background: #007bff; color: white; }

    .buttons button:hover {
      opacity: 0.9;
    }

    ul#laps {
      list-style: none;
      padding: 0;
      margin-top: 20px;
      max-height: 150px;
      overflow-y: auto;
    }

    #laps li {
      background: #e9ecef;
      margin: 5px 0;
      padding: 8px;
      border-radius: 5px;
    }
  </style>
</head>
<body>
  <div class="stopwatch">
    <h1>Stopwatch</h1>
    <div id="display">00:00:00.00</div>
    <div class="buttons">
      <button class="start" onclick="startStop()">Start</button>
      <button class="pause" onclick="pause()">Pause</button>
      <button class="reset" onclick="reset()">Reset</button>
      <button class="lap" onclick="lap()">Lap</button>
    </div>
    <ul id="laps"></ul>
  </div>

  <script>
    let [ms, s, m, h] = [0, 0, 0, 0];
    let display = document.getElementById("display");
    let laps = document.getElementById("laps");
    let timer = null;

    function updateDisplay() {
      let msStr = ms < 10 ? "0" + ms : ms;
      let sStr = s < 10 ? "0" + s : s;
      let mStr = m < 10 ? "0" + m : m;
      let hStr = h < 10 ? "0" + h : h;
      display.innerText = `${hStr}:${mStr}:${sStr}.${msStr}`;
    }

    function stopwatch() {
      ms++;
      if (ms === 100) {
        ms = 0;
        s++;
      }
      if (s === 60) {
        s = 0;
        m++;
      }
      if (m === 60) {
        m = 0;
        h++;
      }
      updateDisplay();
    }

    function startStop() {
      if (timer !== null) return;
      timer = setInterval(stopwatch, 10);
    }

    function pause() {
      clearInterval(timer);
      timer = null;
    }

    function reset() {
      clearInterval(timer);
      timer = null;
      [ms, s, m, h] = [0, 0, 0, 0];
      updateDisplay();
      laps.innerHTML = "";
    }

    function lap() {
      if (timer === null) return;
      let li = document.createElement("li");
      li.innerText = display.innerText;
      laps.appendChild(li);
    }
  </script>
</body>
</html>
```
